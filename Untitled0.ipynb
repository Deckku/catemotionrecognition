```python
import os
import shutil
import random

# Chemin source où les fichiers audio sont stockés
source_dir = '/content/drive/MyDrive/NAYA_DATA_AUG1X'

# Chemins pour les ensembles d'entraînement et de validation
train_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/train'
val_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/val'

# Créer les répertoires s'ils n'existent pas déjà
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Liste des classes
classes = ['Angry', 'Defence', 'Fighting', 'Happy', 'HuntingMind',
           'Mating', 'MotherCall', 'Paining', 'Resting', 'Warning']

# Fonction pour répartir les fichiers audio en ensembles d'entraînement et de validation
def split_data(source_dir, train_dir, val_dir, split_ratio=0.8):
    for class_name in classes:
        # Créer des sous-dossiers pour chaque classe dans train et val
        class_train_dir = os.path.join(train_dir, class_name)
        class_val_dir = os.path.join(val_dir, class_name)
        os.makedirs(class_train_dir, exist_ok=True)
        os.makedirs(class_val_dir, exist_ok=True)

        # Liste des fichiers audio dans chaque classe
        class_dir = os.path.join(source_dir, class_name)
        if not os.path.exists(class_dir):  # Vérifie si le dossier existe
            print(f"[AVERTISSEMENT] Le dossier {class_name} n'existe pas dans {source_dir}.")
            continue

        audio_files = [f for f in os.listdir(class_dir) if os.path.isfile(os.path.join(class_dir, f))]

        # Si aucun fichier n'est trouvé dans le dossier, affiche un avertissement
        if not audio_files:
            print(f"[AVERTISSEMENT] Aucun fichier audio trouvé dans {class_dir}.")
            continue

        random.shuffle(audio_files)  # Mélanger les fichiers

        # Calcul du nombre de fichiers pour l'entraînement
        train_size = int(len(audio_files) * split_ratio)
        train_files = audio_files[:train_size]
        val_files = audio_files[train_size:]

        # Déplacer les fichiers dans les dossiers train et val correspondants
        for file in train_files:
            source_file = os.path.join(class_dir, file)
            destination_file = os.path.join(class_train_dir, file)
            shutil.move(source_file, destination_file)

        for file in val_files:
            source_file = os.path.join(class_dir, file)
            destination_file = os.path.join(class_val_dir, file)
            shutil.move(source_file, destination_file)

        print(f"[INFO] {len(train_files)} fichiers déplacés vers {class_train_dir}")
        print(f"[INFO] {len(val_files)} fichiers déplacés vers {class_val_dir}")

# Appel de la fonction pour organiser les fichiers audio
split_data(source_dir, train_dir, val_dir)
print("Fichiers audio répartis entre les ensembles d'entraînement et de validation.")

```

```python
#Prétraitement et Préparation des Données
import os
import librosa
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def clean_audio(y, sr, low_freq=200, high_freq=8000):
    """
    Filtrer l'audio pour conserver les fréquences correspondant aux vocalisations des chats.
    """
    y_filtered = librosa.effects.preemphasis(y)
    stft = librosa.stft(y_filtered, n_fft=2048, hop_length=512)
    freqs = librosa.fft_frequencies(sr=sr, n_fft=2048)
    mask = (freqs >= low_freq) & (freqs <= high_freq)
    stft_filtered = stft[mask, :]
    y_cleaned = librosa.istft(stft_filtered, hop_length=512)
    return y_cleaned

def extract_spectrogram(y, sr, max_pad_len=128):
    """
    Extraire le spectrogramme en échelle logarithmique pour un signal donné.
    """
    try:
        fmax = sr / 2  # Ajuste fmax pour respecter la règle de Nyquist
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=fmax)
        log_S = librosa.power_to_db(S, ref=np.max)
        pad_width = max_pad_len - log_S.shape[1]
        if pad_width > 0:
            log_S = np.pad(log_S, ((0, 0), (0, pad_width)), mode='constant')
        else:
            log_S = log_S[:, :max_pad_len]
        return log_S
    except Exception as e:
        print(f"[ERREUR] Extraction du spectrogramme échouée : {e}")
        return None

def load_audio_files(base_dir, max_pad_len=128):
    """
    Charger les fichiers audio depuis le répertoire spécifié, nettoyer les audios,
    extraire les spectrogrammes et retourner les données et les étiquettes.
    """
    audio_data = []
    labels = []

    if not os.path.exists(base_dir):
        print(f"[ERREUR] Le répertoire {base_dir} n'existe pas.")
        return np.array([]), []

    print(f"[INFO] Exploration du dossier : {base_dir}")
    for label in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, label)
        if not os.path.isdir(folder_path):
            print(f"[AVERTISSEMENT] {folder_path} n'est pas un dossier, ignoré.")
            continue

        print(f"[INFO] Chargement de la classe : {label}")
        for file in os.listdir(folder_path):
            if file.endswith('.mp3'):
                file_path = os.path.join(folder_path, file)
                try:
                    y, sr = librosa.load(file_path, sr=22050)  # Utiliser une fréquence standard de 22 050 Hz
                    y_cleaned = clean_audio(y, sr)
                    spectrogram = extract_spectrogram(y_cleaned, sr, max_pad_len)
                    if spectrogram is not None and spectrogram.shape == (128, max_pad_len):
                        audio_data.append(spectrogram)
                        labels.append(label)
                    else:
                        print(f"[AVERTISSEMENT] Spectrogramme incorrect pour le fichier : {file}")
                except Exception as e:
                    print(f"[ERREUR] Erreur lors du traitement de {file}: {e}")

    return np.array(audio_data), np.array(labels)

def process_data_for_all_sets(train_dir, val_dir, test_dir, max_pad_len=128):
    """
    Charger les ensembles d'entraînement, validation et test,
    extraire les spectrogrammes, normaliser les données et encoder les labels.
    """
    print("[INFO] Traitement de l'ensemble d'entraînement...")
    X_train, y_train = load_audio_files(train_dir, max_pad_len)

    print("[INFO] Traitement de l'ensemble de validation...")
    X_val, y_val = load_audio_files(val_dir, max_pad_len)

    print("[INFO] Traitement de l'ensemble de test...")
    X_test, y_test = load_audio_files(test_dir, max_pad_len)

    # Vérification des ensembles non vides
    if X_train.size == 0 or len(y_train) == 0:
        raise ValueError("[ERREUR] Aucun fichier audio dans l'ensemble d'entraînement.")
    if X_val.size == 0 or len(y_val) == 0:
        print("[AVERTISSEMENT] Aucun fichier audio chargé dans l'ensemble de validation.")
    if X_test.size == 0 or len(y_test) == 0:
        print("[AVERTISSEMENT] Aucun fichier audio chargé dans l'ensemble de test.")

    # Vérification des labels communs dans tous les ensembles
    all_labels = np.unique(np.concatenate([y_train, y_val, y_test]))
    print(f"[INFO] Labels trouvés dans tous les ensembles : {all_labels}")

    # Calcul des moyennes et écarts-types pour le jeu d'entraînement
    mean = np.mean(X_train, axis=0, keepdims=True)
    std = np.std(X_train, axis=0, keepdims=True)

    # Normalisation des ensembles
    print("[INFO] Normalisation des ensembles...")
    X_train = (X_train - mean) / (std + 1e-8)
    if X_val.size > 0:
        X_val = (X_val - mean) / (std + 1e-8)
    if X_test.size > 0:
        X_test = (X_test - mean) / (std + 1e-8)

    # Encodage des labels
    print("[INFO] Encodage des labels...")
    label_encoder = LabelEncoder()
    label_encoder.fit(np.concatenate([y_train, y_val, y_test]))
    y_train_encoded = label_encoder.transform(y_train)
    y_val_encoded = label_encoder.transform(y_val) if len(y_val) > 0 else []
    y_test_encoded = label_encoder.transform(y_test) if len(y_test) > 0 else []

    return X_train, X_val, X_test, y_train_encoded, y_val_encoded, y_test_encoded, label_encoder

def display_random_spectrograms(data, labels, class_names, num_samples=5):
    """
    Affiche aléatoirement quelques spectrogrammes avec leurs étiquettes correspondantes.
    """
    if len(data) == 0:
        print("[ERREUR] Pas de données disponibles pour l'affichage.")
        return

    indices = random.sample(range(len(data)), min(num_samples, len(data)))
    plt.figure(figsize=(15, 10))

    for i, idx in enumerate(indices):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(data[idx], aspect='auto', origin='lower', cmap='viridis')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f"Classe: {class_names[labels[idx]]}")
        plt.xlabel("Temps")
        plt.ylabel("Fréquences")
        plt.tight_layout()

    plt.show()

# Répertoires des ensembles
train_dir = "/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/train"
val_dir = "/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/val"
test_dir = "/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/test"

# Traitement des données
X_train, X_val, X_test, y_train, y_val, y_test, label_encoder = process_data_for_all_sets(
    train_dir, val_dir, test_dir
)

# Afficher des spectrogrammes aléatoires
print("[INFO] Affichage de spectrogrammes aléatoires de l'ensemble d'entraînement...")
display_random_spectrograms(X_train, y_train, label_encoder.classes_, num_samples=5)

# Affichage des dimensions finales
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")




```

```python
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from tensorflow.keras.regularizers import l2

# Configuration GPU (optionnel)
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)

# Fonction de nettoyage audio
def clean_audio(y, sr, low_freq=200, high_freq=8000):
    y_filtered = librosa.effects.preemphasis(y)
    stft = librosa.stft(y_filtered, n_fft=2048, hop_length=512)
    freqs = librosa.fft_frequencies(sr=sr, n_fft=2048)
    mask = (freqs >= low_freq) & (freqs <= high_freq)
    stft_filtered = stft[mask, :]
    y_cleaned = librosa.istft(stft_filtered, hop_length=512)
    return y_cleaned

# Extraction des spectrogrammes
def extract_spectrogram(y, sr, max_pad_len=128):
    try:
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=sr/2)
        log_S = librosa.power_to_db(S, ref=np.max)
        pad_width = max_pad_len - log_S.shape[1]
        if pad_width > 0:
            log_S = np.pad(log_S, ((0, 0), (0, pad_width)), mode='constant')
        else:
            log_S = log_S[:, :max_pad_len]
        return log_S
    except Exception as e:
        print(f"[ERREUR] Extraction du spectrogramme échouée : {e}")
        return None

# Chargement des données
def load_data(base_dir, yamnet_model=None, use_yamnet=False):
    audio_data = []
    labels = []
    classes = os.listdir(base_dir)

    for label in classes:
        folder_path = os.path.join(base_dir, label)
        if not os.path.isdir(folder_path):
            continue

        print(f"[INFO] Extraction pour la classe : {label}")
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.mp3')]
        if not files:
            print(f"[AVERTISSEMENT] Aucun fichier trouvé dans {folder_path}.")
            continue

        embeddings = []
        for file in files:
            y, sr = librosa.load(file, sr=22050)
            y_cleaned = clean_audio(y, sr)
            spectrogram = extract_spectrogram(y_cleaned, sr)
            if spectrogram is not None:
                embeddings.append(spectrogram)

        audio_data.extend(embeddings)
        labels.extend([label] * len(embeddings))

    return np.array(audio_data), np.array(labels)

# Construction du modèle CNN avec régularisation L2 et Dropout
def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Input(shape=input_shape),

        # Bloc convolutionnel 1
        Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

        # Bloc convolutionnel 2
        Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4),

        # Bloc convolutionnel 3
        Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.5),

        # Couches denses
        Flatten(),
        Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')  # Sortie pour classification
    ])

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Visualisation des performances
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(1, len(acc) + 1)

    plt.plot(epochs, acc, 'bo', label='Précision Entraînement')
    plt.plot(epochs, val_acc, 'b', label='Précision Validation')
    plt.title('Précision Entraînement et Validation')
    plt.legend()
    plt.figure()

    plt.plot(epochs, loss, 'bo', label='Perte Entraînement')
    plt.plot(epochs, val_loss, 'b', label='Perte Validation')
    plt.title('Perte Entraînement et Validation')
    plt.legend()
    plt.show()

# Répertoires des données
train_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/train'
val_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/val'
test_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/test'

# Chargement et extraction des données
X_train, y_train = load_data(train_dir, use_yamnet=False)
X_val, y_val = load_data(val_dir, use_yamnet=False)
X_test, y_test = load_data(test_dir, use_yamnet=False)

# Préparation des données
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
y_test_encoded = label_encoder.transform(y_test)

# Ajout d'une dimension pour les canaux
X_train = X_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]
X_test = X_test[..., np.newaxis]

# Construction du modèle CNN
input_shape = X_train.shape[1:]  # Format (hauteur, largeur, 1)
num_classes = len(label_encoder.classes_)
model = build_cnn_model(input_shape, num_classes)

# Ajout de ReduceLROnPlateau pour ajuster dynamiquement le taux d'apprentissage
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

# Entraînement avec les nouvelles callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train_encoded, validation_data=(X_val, y_val_encoded),
                    epochs=50, batch_size=32, callbacks=[early_stopping, reduce_lr])

# Sauvegarde du modèle
model.save('/content/drive/MyDrive/catemotionrecognitionbyaudio/models/cnn_model.h5')

# Visualisation des performances
plot_training_history(history)

# Évaluation finale
test_loss, test_accuracy = model.evaluate(X_test, y_test_encoded, batch_size=32)
print(f"Test Accuracy: {test_accuracy:.2f}, Test Loss: {test_loss:.2f}")
```

```python
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from tensorflow.keras.regularizers import l2

# Configuration GPU (optionnel)
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)

# Fonction de nettoyage audio
def clean_audio(y, sr, low_freq=200, high_freq=8000):
    y_filtered = librosa.effects.preemphasis(y)
    stft = librosa.stft(y_filtered, n_fft=2048, hop_length=512)
    freqs = librosa.fft_frequencies(sr=sr, n_fft=2048)
    mask = (freqs >= low_freq) & (freqs <= high_freq)
    stft_filtered = stft[mask, :]
    y_cleaned = librosa.istft(stft_filtered, hop_length=512)
    return y_cleaned

# Extraction des spectrogrammes
def extract_spectrogram(y, sr, max_pad_len=128):
    try:
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=sr/2)
        log_S = librosa.power_to_db(S, ref=np.max)
        pad_width = max_pad_len - log_S.shape[1]
        if pad_width > 0:
            log_S = np.pad(log_S, ((0, 0), (0, pad_width)), mode='constant')
        else:
            log_S = log_S[:, :max_pad_len]
        return log_S
    except Exception as e:
        print(f"[ERREUR] Extraction du spectrogramme échouée : {e}")
        return None

# Chargement des données
def load_data(base_dir, yamnet_model=None, use_yamnet=False):
    audio_data = []
    labels = []
    classes = os.listdir(base_dir)

    for label in classes:
        folder_path = os.path.join(base_dir, label)
        if not os.path.isdir(folder_path):
            continue

        print(f"[INFO] Extraction pour la classe : {label}")
        files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.mp3')]
        if not files:
            print(f"[AVERTISSEMENT] Aucun fichier trouvé dans {folder_path}.")
            continue

        embeddings = []
        for file in files:
            y, sr = librosa.load(file, sr=22050)
            y_cleaned = clean_audio(y, sr)
            spectrogram = extract_spectrogram(y_cleaned, sr)
            if spectrogram is not None:
                embeddings.append(spectrogram)

        audio_data.extend(embeddings)
        labels.extend([label] * len(embeddings))

    return np.array(audio_data), np.array(labels)

# Construction du modèle CNN avec régularisation L2 et Dropout
def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Input(shape=input_shape),

        # Bloc convolutionnel 1
        Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

        # Bloc convolutionnel 2
        Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4),

        # Bloc convolutionnel 3
        Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.5),

        # Couches denses
        Flatten(),
        Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')  # Sortie pour classification
    ])

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Visualisation des performances
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(1, len(acc) + 1)

    plt.plot(epochs, acc, 'bo', label='Précision Entraînement')
    plt.plot(epochs, val_acc, 'b', label='Précision Validation')
    plt.title('Précision Entraînement et Validation')
    plt.legend()
    plt.figure()

    plt.plot(epochs, loss, 'bo', label='Perte Entraînement')
    plt.plot(epochs, val_loss, 'b', label='Perte Validation')
    plt.title('Perte Entraînement et Validation')
    plt.legend()
    plt.show()

# Répertoires des données
train_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/train'
val_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/val'
test_dir = '/content/drive/MyDrive/catemotionrecognitionbyaudio/datasets/test'

# Chargement et extraction des données
X_train, y_train = load_data(train_dir, use_yamnet=False)
X_val, y_val = load_data(val_dir, use_yamnet=False)
X_test, y_test = load_data(test_dir, use_yamnet=False)

# Préparation des données
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
y_test_encoded = label_encoder.transform(y_test)

# Ajout d'une dimension pour les canaux
X_train = X_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]
X_test = X_test[..., np.newaxis]

# Construction du modèle CNN
input_shape = X_train.shape[1:]  # Format (hauteur, largeur, 1)
num_classes = len(label_encoder.classes_)
model = build_cnn_model(input_shape, num_classes)

# Ajout de ReduceLROnPlateau pour ajuster dynamiquement le taux d'apprentissage
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

# Entraînement avec les nouvelles callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train_encoded, validation_data=(X_val, y_val_encoded),
                    epochs=50, batch_size=32, callbacks=[early_stopping, reduce_lr])

# Sauvegarde du modèle
model.save('/content/drive/MyDrive/catemotionrecognitionbyaudio/models/cnn_model.h5')

# Visualisation des performances
plot_training_history(history)

# Évaluation finale
test_loss, test_accuracy = model.evaluate(X_test, y_test_encoded, batch_size=32)
print(f"Test Accuracy: {test_accuracy:.2f}, Test Loss: {test_loss:.2f}")


```
